3. Considere la serie convergente

  $\frac{π}{4}=1-\frac{1}{3}+\frac{1}{5}-\frac{1}{7}+...$

  - La serie es absolutamente convergente o condicionalmente convergente? Justifique su respuesta de forma rigurosa.

  La serie se puede reescribir como
  $\frac{π}{4}= \sum_{n=0}^∞ (-1)^n \frac{1}{2n+1}$
  Para saber si la serie es *absolutamente convergente* evaluamos los valores absolutos
  $\frac{π}{4}= \sum_{n=0}^∞ |(-1)^n \frac{1}{2n+1}| = \sum_{n=0}^∞ \frac{1}{2n+1}$
  Teniendo en cuenta que la serie es un subconjunto con los impares de la serie $\sum_{n=0}^∞ \frac{1}{n}$ podemos inducir que esta serie al igual que la serie armónica ($\sum_{n=0}^∞ \frac{1}{n}$) es divergente.
  La serie original es convergente hacia $\frac{π}{4}$, esto lo sabemos gracias a la teoría de series alternantes y a la fórmula de Leibniz para el cálculo de π, por lo que la serie es *condicionalmente convergente*.
  - Escriba un programa en Python que calcule la suma parcial ¿Cuántos términos son necesarios para obtener 3 dígitos correctos de $\frac{π}{4}$?



```python
def calcular_pi_sobre_4(precision=0.0001):
    suma = 0.0
    termino = 1.0
    n = 0
    signo = 1

    while abs(termino) >= precision:
        termino = signo * 1.0 / (2 * n + 1)
        suma += termino
        signo *= -1
        n += 1

    return suma, n

# Calculamos π/4 con una precisión de tres dígitos decimales
pi_sobre_4, num_terminos = calcular_pi_sobre_4(0.0001)

print(f"Aproximación de π/4: {pi_sobre_4}")
print(f"Número de términos necesarios: {num_terminos}")
print(f"π/4 real aproximado: {3.141592653589793 / 4}")
```

```python
def decimal_a_binario(decimal):
    binario = ""
    while decimal > 0:
        binario = str(decimal % 2) + binario
        decimal //= 2
    return binario

print(f"23 = {decimal_a_binario(23)}")
print(f"87 = {decimal_a_binario(87)}")
print(f"378 = {decimal_a_binario(378)}")
```

```python
# Pasar del decimal a la mantiza
def fraccion_a_binario(f, limite=20):
    """ Convierte la fracción decimal a binario con un límite de dígitos. """
    binario = ''
    count = 0
    while f != 0 and count < limite:
        f *= 2
        bit = int(f)
        binario += str(bit)
        f -= bit
        count += 1
    return binario

def decimal_a_mantisa(num):
    """ Convierte un número decimal a su mantisa en notación científica binaria. """
    if num < 0:
        signo = -1
        num = -num
    else:
        signo = 1

    # Dividir el número en su parte entera y fraccionaria
    parte_entera = int(num)
    fraccion = num - parte_entera

    # Convertir ambas partes a binario
    binario_entero = decimal_a_binario(parte_entera)
    binario_fraccion = fraccion_a_binario(fraccion)

    # Normalizar el número binario (1.bbbb forma)
    if parte_entera != 0:
        # Encontrar la primera posición del '1' en la parte entera
        exponente = len(binario_entero) - 1
        mantisa = binario_entero[1:] + binario_fraccion
    else:
        # Si la parte entera es 0, encontrar el primer '1' en la fracción
        primer_uno = binario_fraccion.find('1')
        exponente = -(primer_uno + 1)
        mantisa = binario_fraccion[primer_uno + 1:]

    return signo, mantisa, exponente

numero_decimal = 23
signo, mantisa, exponente = decimal_a_mantisa(numero_decimal)
print(f"Signo: {'+' if signo == 1 else '-'}")
print(f"Mantisa: {mantisa}")
print(f"Exponente: {exponente}")
signo, mantisa, exponente = decimal_a_mantisa(87)
print(f"Signo: {'+' if signo == 1 else '-'}")
print(f"Mantisa: {mantisa}")
print(f"Exponente: {exponente}")
signo, mantisa, exponente = decimal_a_mantisa(378)
print(f"Signo: {'+' if signo == 1 else '-'}")
print(f"Mantisa: {mantisa}")
print(f"Exponente: {exponente}")
```

5. Con ayuda de la convergencia de la serie geométrica halle una representación como una serie para la siguiente función, además entregue el radio de convergencia.

  $f(x) = \frac{1}{1-x^2}$

  Teniendo en cuenta

  $\frac{1}{1-x} = \sum_{n=0}^∞ x^n$, para $|x| < 1$

  Si sustituímos $x^2$ por $x$ tendríamos:

  $\frac{1}{1-(x^2)}= \sum_{n=0}^∞ (x^2)^n = \sum_{n=0}^∞ x^{2n}$, para $|x^2| < 1$

  Por lo que la representación como una serie sería:

  $f(x)= \sum_{n=0}^∞ x^{2n}$

  Para determinar el radio de convergencia se analiza el límite superior para $|x|$ tal que $|x| < 1$ , podemos aplicar el criterio de la raíz:

  $lim_{n → ∞} \sqrt[2n]{|x^{2n}|} = lim_{n →∞} |x| = 1$

  Por lo que tiene un radio de convergencia de 1
