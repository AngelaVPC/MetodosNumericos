```python
# Spline para funcioens lineales
import numpy as np

def spline_lineal(x, y, x0):
    n = len(x)
    for i in range(n-1):
        if x[i] <= x0 <= x[i+1]:
            return y[i] + (y[i+1] - y[i])/(x[i+1] - x[i])*(x0 - x[i])
        
    return None

def graph_spline_lineal(x, y, x0):
    import matplotlib.pyplot as plt
    x0 = np.linspace(x[0], x[-1], 100)
    y0 = [spline_lineal(x, y, xi) for xi in x0]
    plt.plot(x, y, 'ro')
    plt.plot(x0, y0)
    plt.show()
    
```

```python
puntos = [(0, 0), (1, 1), (2, 0), (4, 10)]
x = [p[0] for p in puntos]
y = [p[1] for p in puntos]
x0 = 1.5
print(spline_lineal(x, y, x0))
graph_spline_lineal(x, y, x0)
```
